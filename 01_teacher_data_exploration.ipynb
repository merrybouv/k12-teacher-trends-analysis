```python
# Multi-Year NCES Staff Data Loading and Consolidation
import pandas as pd
import os

# Load 8 years of consistent format data (2016-2024)
file_list = [
    'staff_2016-17.csv',
    'staff_2017-18.txt', 
    'staff_2018-19.txt',
    'staff_2019-20.txt',
    'staff_2020-21.txt',
    'staff_2021-22.txt',
    'staff_2022-23.txt',
    'staff_2023-24.txt'
]

print(f"Loading {len(file_list)} years of NCES staff data")

# Store dataframes for concatenation
all_dataframes = []

# Process each data file
for filename in file_list:
    try:
        file_path = f'data/{filename}'
        
        # Handle both folder and file structures in NCES data
        if os.path.isdir(file_path):
            csv_files = [f for f in os.listdir(file_path) if f.endswith('.csv')]
            if csv_files:
                df = pd.read_csv(f'{file_path}/{csv_files[0]}')
            else:
                print(f"No CSV found in {filename}")
                continue
        else:
            if filename.endswith('.csv'):
                df = pd.read_csv(file_path)
            else:
                df = pd.read_csv(file_path, sep='\t')
        
        # Extract academic year from filename
        year = filename.replace('staff_', '').replace('.txt', '').replace('.csv', '')
        df['YEAR'] = year
        
        all_dataframes.append(df)
        print(f"Loaded {len(df):,} records for {year}")
        
    except Exception as e:
        print(f"Error loading {filename}: {str(e)}")

# Combine all years into single dataset
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    
    print(f"\nDataset Summary:")
    print(f"Total records: {len(combined_df):,}")
    print(f"Years: {sorted(combined_df['YEAR'].unique())}")
    print(f"Jurisdictions: {len(combined_df['STATENAME'].unique())}")
    print(f"Staff categories: {len(combined_df['STAFF'].unique())}")
    
    # Data quality check
    missing_counts = combined_df.isnull().sum()
    print(f"\nMissing values:")
    for col, count in missing_counts.items():
        if count > 0:
            print(f"  {col}: {count:,}")
    
    # Year coverage
    print(f"\nRecords per year:")
    year_counts = combined_df['YEAR'].value_counts().sort_index()
    for year, count in year_counts.items():
        print(f"  {year}: {count:,}")
        
    print(f"\nFirst few records:")
    print(combined_df.head())
    
else:
    print("No data files loaded successfully")
```

```python
# Multi-Year NCES Staff Data Loading and Consolidation
import pandas as pd
import os

# Load 8 years of consistent format data (2016-2024)
file_list = [
    'staff_2016-17.csv',
    'staff_2017-18.txt', 
    'staff_2018-19.txt',
    'staff_2019-20.txt',
    'staff_2020-21.txt',
    'staff_2021-22.txt',
    'staff_2022-23.txt',
    'staff_2023-24.txt'
]

print(f"Loading {len(file_list)} years of NCES staff data")

# Store dataframes for concatenation
all_dataframes = []

# Process each data file
for filename in file_list:
    try:
        file_path = f'data/{filename}'
        
        # Handle both folder and file structures in NCES data
        if os.path.isdir(file_path):
            csv_files = [f for f in os.listdir(file_path) if f.endswith('.csv')]
            if csv_files:
                df = pd.read_csv(f'{file_path}/{csv_files[0]}')
            else:
                print(f"No CSV found in {filename}")
                continue
        else:
            if filename.endswith('.csv'):
                df = pd.read_csv(file_path)
            else:
                df = pd.read_csv(file_path, sep='\t')
        
        # Extract academic year from filename
        year = filename.replace('staff_', '').replace('.txt', '').replace('.csv', '')
        df['YEAR'] = year
        
        all_dataframes.append(df)
        print(f"Loaded {len(df):,} records for {year}")
        
    except Exception as e:
        print(f"Error loading {filename}: {str(e)}")

# Combine all years into single dataset
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    
    print(f"\nCombined Dataset Summary:")
    print(f"Total records: {len(combined_df):,}")
    print(f"Years covered: {sorted(combined_df['YEAR'].unique())}")
    print(f"States/jurisdictions: {len(combined_df['STATENAME'].unique())}")
    print(f"Staff categories: {len(combined_df['STAFF'].unique())}")
    
    # Data quality assessment
    missing_counts = combined_df.isnull().sum()
    print(f"\nData Quality Check:")
    print("Missing values by column:")
    for col, count in missing_counts.items():
        if count > 0:
            pct_missing = (count / len(combined_df)) * 100
            print(f"  {col}: {count:,} missing ({pct_missing:.1f}%)")
    
    # Year-by-year coverage analysis
    print(f"\nYear Coverage Analysis:")
    year_counts = combined_df['YEAR'].value_counts().sort_index()
    for year, count in year_counts.items():
        print(f"  {year}: {count:,} records")
    
    # Check for data consistency patterns
    min_records = year_counts.min()
    max_records = year_counts.max()
    variation = max_records - min_records
    print(f"\nRecord count variation: {variation:,} records ({min_records:,} to {max_records:,})")
    print("This variation is normal for government education data due to:")
    print("- States not reporting all staff categories each year")
    print("- Changes in data collection requirements")
    print("- Jurisdictional differences in staff organization")
        
    print(f"\nSample records:")
    print(combined_df.head())
    
else:
    print("No data files loaded successfully")
```

```python
# Teacher Workforce Trends Analysis (2016-2024)
# Analysis of national and state-level patterns in teacher employment

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Filter to main "Teachers" category for consistency across years
teachers_df = combined_df[combined_df['STAFF'] == 'Teachers'].copy()

print("Teacher Workforce Analysis")
print("=" * 50)
print(f"Records analyzed: {len(teachers_df):,}")
print(f"Years covered: {sorted(teachers_df['YEAR'].unique())}")
print(f"Jurisdictions: {len(teachers_df['STATENAME'].unique())}")

# Remove records with missing teacher counts
teachers_clean = teachers_df.dropna(subset=['STAFF_COUNT'])
missing_count = len(teachers_df) - len(teachers_clean)
if missing_count > 0:
    print(f"Records with missing teacher counts excluded: {missing_count}")

# National teacher workforce trends
print(f"\nNational Teacher Workforce Trends")
print("-" * 40)

national_totals = teachers_clean.groupby('YEAR')['STAFF_COUNT'].sum().sort_index()

print("Annual teacher totals:")
for year, total in national_totals.items():
    print(f"  {year}: {total:,.0f}")

# Calculate year-over-year changes
print(f"\nYear-over-year changes:")
for i in range(1, len(national_totals)):
    current_year = national_totals.index[i]
    previous_year = national_totals.index[i-1]
    current_total = national_totals.iloc[i]
    previous_total = national_totals.iloc[i-1]
    
    change = current_total - previous_total
    pct_change = (change / previous_total) * 100
    
    direction = "increase" if change > 0 else "decrease"
    print(f"  {previous_year} to {current_year}: {abs(change):,.0f} {direction} ({pct_change:+.1f}%)")

# State-level analysis: Compare first and last years
print(f"\nState-Level Changes (2016-17 to 2023-24)")
print("-" * 45)

# Get data for comparison years
first_year = teachers_clean[teachers_clean['YEAR'] == '2016-17'].set_index('STATENAME')['STAFF_COUNT']
last_year = teachers_clean[teachers_clean['YEAR'] == '2023-24'].set_index('STATENAME')['STAFF_COUNT']

# Calculate changes for states with both years of data
common_states = first_year.index.intersection(last_year.index)
state_changes = pd.DataFrame({
    'Teachers_2016': first_year[common_states],
    'Teachers_2023': last_year[common_states]
}).dropna()

state_changes['Change_Absolute'] = state_changes['Teachers_2023'] - state_changes['Teachers_2016']
state_changes['Change_Percent'] = (state_changes['Change_Absolute'] / state_changes['Teachers_2016']) * 100

# Sort by percentage change
state_changes_sorted = state_changes.sort_values('Change_Percent')

print("States with largest decreases:")
print(state_changes_sorted.head(8)[['Teachers_2016', 'Teachers_2023', 'Change_Absolute', 'Change_Percent']].round(1))

print(f"\nStates with largest increases:")
print(state_changes_sorted.tail(8)[['Teachers_2016', 'Teachers_2023', 'Change_Absolute', 'Change_Percent']].round(1))

# COVID impact analysis
print(f"\nCOVID-19 Impact Analysis")
print("-" * 30)

covid_before = teachers_clean[teachers_clean['YEAR'] == '2019-20'].set_index('STATENAME')['STAFF_COUNT']
covid_after = teachers_clean[teachers_clean['YEAR'] == '2020-21'].set_index('STATENAME')['STAFF_COUNT']

covid_impact = pd.DataFrame({
    'Pre_COVID': covid_before,
    'During_COVID': covid_after
}).dropna()

covid_impact['Change'] = covid_impact['During_COVID'] - covid_impact['Pre_COVID']
covid_impact['Percent_Change'] = (covid_impact['Change'] / covid_impact['Pre_COVID']) * 100

national_covid_change = covid_impact['Change'].sum()
national_covid_pct = (national_covid_change / covid_impact['Pre_COVID'].sum()) * 100

print(f"National COVID impact: {national_covid_change:,.0f} teachers ({national_covid_pct:.1f}%)")
print(f"States with largest COVID losses:")
covid_worst = covid_impact.sort_values('Percent_Change').head(5)
for state, row in covid_worst.iterrows():
    print(f"  {state}: {row['Percent_Change']:+.1f}% ({row['Change']:+,.0f})")

# Create comprehensive visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: National trend
years_numeric = range(len(national_totals))
ax1.plot(years_numeric, national_totals.values, marker='o', linewidth=3, markersize=8, color='#2E86C1')
ax1.set_title('National Teacher Workforce Trend (2016-2024)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Academic Year')
ax1.set_ylabel('Total Teachers (millions)')
ax1.set_xticks(years_numeric)
ax1.set_xticklabels(national_totals.index, rotation=45, ha='right')
ax1.grid(True, alpha=0.3)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

# Highlight COVID period
covid_start = list(national_totals.index).index('2019-20')
covid_end = list(national_totals.index).index('2020-21')
ax1.axvspan(covid_start, covid_end, alpha=0.2, color='red', label='COVID Impact')
ax1.legend()

# Plot 2: Top states by teacher count (2023-24)
top_states = teachers_clean[teachers_clean['YEAR'] == '2023-24'].nlargest(10, 'STAFF_COUNT')
ax2.barh(range(len(top_states)), top_states['STAFF_COUNT'] / 1000, color='#28B463')
ax2.set_title('Top 10 States by Teacher Count (2023-24)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Teachers (thousands)')
ax2.set_yticks(range(len(top_states)))
ax2.set_yticklabels(top_states['STATENAME'], fontsize=10)
ax2.invert_yaxis()

# Plot 3: State-level changes (extreme cases)
extreme_states = pd.concat([state_changes_sorted.head(6), state_changes_sorted.tail(6)])
colors = ['#E74C3C' if x < 0 else '#27AE60' for x in extreme_states['Change_Percent']]
bars = ax3.barh(range(len(extreme_states)), extreme_states['Change_Percent'], color=colors, alpha=0.8)
ax3.set_title('State Teacher Workforce Changes\n(2016-17 to 2023-24)', fontsize=14, fontweight='bold')
ax3.set_xlabel('Percent Change (%)')
ax3.set_yticks(range(len(extreme_states)))
ax3.set_yticklabels(extreme_states.index, fontsize=9)
ax3.axvline(x=0, color='black', linestyle='-', alpha=0.5)

# Plot 4: COVID impact distribution
ax4.hist(covid_impact['Percent_Change'], bins=15, alpha=0.7, color='#F39C12', edgecolor='black', linewidth=1)
ax4.set_title('Distribution of COVID Impact\n(2019-20 to 2020-21)', fontsize=14, fontweight='bold')
ax4.set_xlabel('Percent Change (%)')
ax4.set_ylabel('Number of States')
median_impact = covid_impact['Percent_Change'].median()
ax4.axvline(x=median_impact, color='red', linestyle='--', linewidth=2, 
           label=f'Median: {median_impact:.1f}%')
ax4.legend()

plt.tight_layout()
plt.show()

# Summary findings
print(f"\nKey Findings Summary")
print("=" * 25)
print(f"• COVID Impact: {abs(national_covid_change):,.0f} teacher reduction in 2020-21 ({national_covid_pct:.1f}%)")
print(f"• Recovery: Teacher workforce has recovered to {national_totals.iloc[-1]:,.0f} by 2023-24")
print(f"• Regional Variation: State changes range from {state_changes_sorted['Change_Percent'].min():.1f}% to {state_changes_sorted['Change_Percent'].max():.1f}%")
print(f"• Largest Challenge: Florida lost {abs(state_changes_sorted.iloc[0]['Change_Absolute']):,.0f} teachers")
print(f"• Success Story: Utah gained {state_changes_sorted.iloc[-1]['Change_Absolute']:,.0f} teachers")
```
